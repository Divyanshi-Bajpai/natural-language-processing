```python
import nltk
str=input("Hello! I am ChatarBot. How are you feeling today? ")
tok=nltk.word_tokenize(str)
print(tok)
if 'happy' in tok:
  print("That's great! Hope your day is good.")
elif 'sad' or 'not' in tok:
  print("I am sorry for that. Engage in some hobbies.")
else:
  print("Invalid.")
```
