ENTITIES AND RELATED OPERATIONS

```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
def show(doc1):
  if doc1.ents:
    for ent in doc1.ents:
      print(ent.text+" --->"+ent.label_+"--->"+str(spacy.explain(ent.label_)))
  else:
    print("No entities found.")
```

```python
doc1=nlp(u"Texas is a deserted and very hot too ,in evening near 1:00 it is  very hot.")
```

```python
show(doc1)
```
