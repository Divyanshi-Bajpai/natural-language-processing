```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp("  We're moving to L.A. !!   ")
for tokens in doc:
  print(tokens.text)
```

While encountering a website or an email, spacy doesn't break it up into more parts and keeps it as it is. In order to find the number of tokoens we use the len function

```python
doc2=nlp("We're happy to help you. In case on any query please contact us on- suport@email.com and reach out to us at http://www.helponyourway.com!!")
for tokens in doc2:
  print(tokens)
len(doc2)
```

```python
doc3=nlp("It's a nice and sunny day. We should go out and have fun!")
for tokens in doc3:
  print(tokens, end=" | ")
```

.ents helps us in finding the entities or in other terms important words having a deep imapce on the meaning of the sentence. Also .label_ helps us in knowing the kind of entity. The last line helps us in explaining the kind of entity changing it into a string and then print it out

```python
doc4=nlp("Apple to build a Hong Kong factory for $6 million")
for entity in doc4.ents:
  print(entity)
  print(entity.label_)
  print(str(spacy.explain(entity.label_)))
  print()
```

```python
from spacy import displacy
doc=nlp("Apple to build a Hong Kong factory for $6 million")
displacy.render(doc,style='dep',jupyter=True,options={'distance':110})
```
